```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import pearsonr as r
```

```python
X1 = pd.read_csv("all_features_QF_SLIA.csv")
y1 = pd.read_csv("times_QF_SLIA_ALL.csv")
X = X1.values
y = np.squeeze(y1.values)

header = X1.columns
```

```python
# For each feature, compute Pearson's R and the corresponding p-value
# If the p-value indicates statistical significance (p <= 0.05), print it
for i in range(X.shape[1]):
    R, P = r(X[:,i],y)
    if P <= 0.05:
        print("Feature " + str(i) + ": (" + header[i] + ") r=" + str(R) + ", p=" + str(P))
```

```python
# Where i is the feature index
def show_correlation_graph(i):
    x = X[:,i]
    plt.scatter(x,y)
    z = np.polyfit(x, y, 1)
    p = np.poly1d(z)
    plt.plot(x,p(x),"r-")
```

```python
show_correlation_graph(12)
```
